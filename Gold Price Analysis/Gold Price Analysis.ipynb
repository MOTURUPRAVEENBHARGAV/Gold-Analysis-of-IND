## Arima Model

```python
from statsmodels.tsa.arima_model import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

```python
X=df["Price(24 karat per 10 grams)"].values
print(X.shape)
train=X[0:70]
test=X[70:]
print(train.shape)
print(test.shape)
```

```python
#p,d,q  
#p=Periods taken for autoregressive model
#d= integrated order
#q-> periods in moving average model
model_arima=ARIMA(train,order=(3,1,1)).fit()
```

```python
predictions=model_arima.forecast(steps=10)[0]
predictions
```

```python
plt.plot(test,color='black')
plt.plot(predictions,color='r')
```

```python
### AIC is a method to choose the best parameters
```

```python
print(model_arima.aic)
```

```python
## lets choose the best p,d,q values
```

```python
import itertools
import operator
p=d=q=range(0,5)
pdq=list(itertools.product(p,d,q))
pdq
```

```python
l=[]
b=[]
for param in pdq:
    try:
        model_arima=ARIMA(train,order=(param)).fit()
        l.append(model_arima.aic)
        b.append(param)
        
        print(model_arima.aic,param)
        
    except:
        continue
print()
a=zip(l,b)
print("Minimum AIC value:",min(a))
```

```python
model_arima=ARIMA(train,order=(2,2,1)).fit()
predictions=model_arima.forecast(steps=25)[0]
predictions
```

```python
plt.plot(test,color='black')
plt.plot(predictions,color='r')
plt.show()
```

```python
### performance metrics
mean_squared_error(test,predictions)
```
